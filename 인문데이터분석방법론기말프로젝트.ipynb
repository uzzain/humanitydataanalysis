```python
import pandas as pd
```

```python
# CSV 파일 읽기
df = pd.read_csv("language_learning_dataset_complete.csv")

# 데이터프레임 정보 확인
print(df.info())
```

```python
# 모국어가 "Chinese" 또는 "Other"인 행 제거
df = df[(df["Native Language"] != "Chinese") & (df["Native Language"] != "Other")]

# 결과 확인
print(df.head())
print(f"모국어가 Chinese 또는 Other가 아닌 행의 개수: {len(df)}")
```

```python
df.head()
```

```python
# 거주 국가가 미국(United States) 또는 중국(China)인 행만 필터링
df = df[(df["Country of Residence"] == "USA") | (df["Country of Residence"] == "China")]

# 결과 확인
print(df.head())
print(f"거주국가가 미국 또는 중국인 행의 개수: {len(df)}")
```

```python
df.tail()
```

```python
columns_to_keep = [
    # 종속변수
    "Speaking Fluency Scores",

    # 독립변수
    "Age",
    "Gender",
    "Native Language",
    "Country of Residence",
    "Language Proficiency Level (Initial)",
    "Type of Activity",
    "Lesson Content Type",
    "Number of Lessons Completed",
    "Time Spent on Learning (Hours)",
    "Learning Platform or Tool Used",
    "Homework Completion Rate (%)",
    "Participation in Interactive Exercises",
    "Frequency of Practice Sessions",
    "Error Correction Rate (%)",
    "Motivation Level",
    "Peer Interaction Score",
    "Engagement with Additional Materials",
    "Learning Environment",
    "Learning Mode",
    "Use of AI Tools"
]

# 필요한 컬럼만 선택하여 데이터프레임 재구성
df = df[columns_to_keep]

# 변경된 데이터프레임 확인
print(df.head())

# 데이터프레임 정보 확인
print(df.info())
```

```python
# 컬럼명의 공백을 "_"로 대체
df.columns = df.columns.str.replace(" ", "_")
```
