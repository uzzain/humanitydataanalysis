```python
def tokenize_and_filter(text):
    # OKT를 사용해서 명사 추출하고 의존명사와 같은 짧은 단어를 필터링
    nouns = okt.nouns(text)
    return [word for word in nouns if len(word) > 1]  # 한글자인 단어 지우기
```

```python
#컬럼 속 공백을 _로 대체
df.columns = df.columns.str.replace(' ', '_')

df.columns
```

```python
# 한국어 명사 추출
okt = Okt()

def extract_nouns(text):
    nouns = okt.nouns(text)
    return ' '.join(nouns)

# 명사만 추출한 뒤 컬럼에 적용
df['제목_명사'] = df['제목'].apply(extract_nouns)
df['신청_데이터_명사'] = df['신청_데이터명'].apply(extract_nouns)

# 클러스터링을 위해 ' '을 구분점으로 두 컬럼을 합치기
df['혼합된_명사'] = df['제목_명사'] + ' ' + df['신청_데이터_명사']

# TF-IDF Vectorization을 이용하여 텍스트를 수치화
vectorizer = TfidfVectorizer(max_features=1000)
X = vectorizer.fit_transform(df['혼합된_명사'])
```

```python
# 제목열과 신청데이터열을 명사로만 적용
df['제목_명사'] = df['제목'].apply(extract_nouns)
df['신청_데이터_명사'] = df['신청_데이터명'].apply(extract_nouns)

# 클러스터링을 위한 신청 데이터와 제목 데이터명사 합치기
df['혼합된_명사'] = df['제목_명사'] + ' ' + df['신청_데이터_명사']
```

```python
#  TF-IDF Vectorization를 사용해서 텍스트를 수치형으로 전환
vectorizer = TfidfVectorizer(max_features=1000)
X = vectorizer.fit_transform(df['혼합된_명사'])
```

```python
# df에 Token 컬럼을 만들어 토큰화된 명사 따로 정리
df['Tokens'] = df['혼합된_명사'].apply(tokenize_and_filter)
```

```python
# 텍스트 내 빈도 세기
word_counts = Counter(all_text.split())

# 불용어 정의
excluded_words = {'현황', '데이터', '관련', '문의', '조회', '엑셀', '등록', '신청',
                  '분석', '자료', '파일', '요청', '정보', '코드', '상세', '목록',
                  '전국', '품목', '가격', '최근', '통계', '월별', '일별', '별'}

# 각 단어별로 5까지 최대 빈도수 제한, 불용어 제거
limited_word_counts = {word: min(count, 5) for word, count in word_counts.items() if word not in excluded_words}

# 수정된 빈도수를 이용하여 워드클라우드 만들기
wordcloud = WordCloud(
    font_path='/content/drive/MyDrive/인문데이터분석/NanumBarunGothic.ttf',
    width=800,
    height=400,
    background_color='white',
    colormap='RdPu'
).generate_from_frequencies(limited_word_counts)

# 워드 클라우드 그리기
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Wordcloud made with extracted nouns 2', fontsize=16)
plt.show()
```

```python
# 데이터 준비
# 'Tokens' 컬럼이 존재하며, 토큰화된 텍스트를 포함하는지 확인
if 'Tokens' not in df.columns:
    raise ValueError("The 'Tokens' column is missing. Ensure you preprocess and tokenize the text data.")

# 불용어 제거하기
df['Tokens'] = df['Tokens'].apply(lambda tokens: [word for word in tokens if word not in excluded_words])

# 딕셔너리와 코퍼스 생성(콜포라?)
dictionary = corpora.Dictionary(df['Tokens'])  # 단어들을 고유 ID로 매칭
corpus = [dictionary.doc2bow(tokens) for tokens in df['Tokens']]  # 문서를 Bag-of-Words 방식으로 표현(빈도수 표)

# LDA 모델 훈련시키기
num_topics = 5  # 필요한 토픽의 수를 조절
lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics, passes=10, random_state=42)

# 키워드와 함께 토픽 출력
print("Topics and their keywords:")
for idx, topic in lda_model.show_topics(num_topics=num_topics, num_words=10, formatted=False):
    print(f"Topic {idx + 1}: ", [word for word, _ in topic])

# 토픽들 시각화
# 시각화 데이터 준비하기
pyLDAvis.enable_notebook()  # 쥬피터 노트북에서도 lda 모델 시각화 할 수 있게
lda_vis = pyLDAvis.gensim_models.prepare(lda_model, corpus, dictionary)

# 시각화하기
pyLDAvis.display(lda_vis)
```

```python
# 클러스터링 레이블 가져오기
labels = df['Cluster']

# 실루엣 점수 계산
silhouette = silhouette_score(X, labels)  # X는 TF-IDF 벡터화된 데이터
print(f"Silhouette Score: {silhouette:.4f}")

# 데이비스-볼딘 인덱스 계산
db_index = davies_bouldin_score(X.toarray(), labels)  # 이 지표는 밀집 배열(dense array) 입력을 필요로 함
print(f"Davies-Bouldin Index: {db_index:.4f}")

# 관성(Inertia) 값 출력 (이미 KMeans 모델에서 제공됨)
#**관성(Inertia)**는 클러스터 내부의 응집도를 측정합니다. 즉, 각 클러스터 내에서 데이터 포인트들이 얼마나 서로 가까운지를 나타냅니다.
# 관성 값이 낮을수록, 데이터 포인트들이 클러스터의 중심에 더 가까운 상태로 모여 있다는 의미입니다. 반대로, 관성 값이 높으면 클러스터들이 더 분산되어 있다는 뜻입니다.
print(f"Inertia: {kmeans.inertia_:.4f}")
```

```python
# 일관성 점수(높을 수록 좋음)
coherence_model = CoherenceModel(model=lda_model, texts=df['Tokens'], dictionary=dictionary, coherence='c_v')
coherence_score = coherence_model.get_coherence()
print(f"Coherence Score: {coherence_score:.4f}")

# 복잡도 점수
perplexity = lda_model.log_perplexity(corpus)
print(f"Perplexity: {perplexity:.4f}")
```

```python
# Step 1: 딕셔너리 재구축
# 문서 빈도 필터링: 최소 5번 이상 등장하고, 문서의 50% 이상에 등장하지 않는 단어만 남김
dictionary.filter_extremes(no_below=5, no_above=0.5)
# 필터링 이후 딕셔너리를 정리하여 효율적으로 관리
dictionary.compactify()

# Step 2: 업데이트된 딕셔너리를 사용하여 코퍼스(corpus) 재생성
# 각 문서의 토큰을 BOW(Bag of Words) 형식으로 변환
# 텍스트 데이터를 LDA(Latent Dirichlet Allocation) 모델이 처리할 수 있는 수치 데이터로 변환
corpus = [dictionary.doc2bow(tokens) for tokens in df['Tokens']]

# Step 3: LDA(Latent Dirichlet Allocation) 모델 학습

lda_model = LdaModel(
    corpus=corpus,             # 코퍼스 입력
    id2word=dictionary,        # 단어-인덱스 맵핑 딕셔너리
    num_topics=5,              # 토픽의 개수 설정
    alpha='auto',              # alpha 하이퍼파라미터 자동 튜닝
    eta='auto',                # eta 하이퍼파라미터 자동 튜닝
    passes=20                  # 코퍼스를 반복 학습할 횟수
)

# 학습된 LDA 모델에서 추출된 토픽 출력
for idx, topic in lda_model.print_topics(num_words=10):  # 각 토픽별 상위 10개의 단어 표시
    print(f"Topic {idx + 1}: {topic}")

# LDA 모델을 추가적으로 학습
lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=5, passes=50)


# 일관성 점수 계산
coherence_model = CoherenceModel(model=lda_model, texts=df['Tokens'], dictionary=dictionary, coherence='c_v')
coherence_score = coherence_model.get_coherence()  # 일관성 점수 계산
print(f"일관성 점수: {coherence_score:.4f}")

# Perplexity(혼란도) 계산
perplexity = lda_model.log_perplexity(corpus)
print(f"응집도 점수: {perplexity:.4f}")
```

```python
# 클러스터링 레이블 가져오기
labels = df['Cluster']

# 실루엣 점수 계산
silhouette = silhouette_score(X, labels)  # X는 TF-IDF 벡터화된 데이터
print(f"Silhouette Score: {silhouette:.4f}")

# 데이비스-볼딘 인덱스 계산
db_index = davies_bouldin_score(X.toarray(), labels)  # 이 지표는 밀집 배열(dense array) 입력을 필요로 함
print(f"Davies-Bouldin Index: {db_index:.4f}")

# 관성(Inertia) 값 출력 (이미 KMeans 모델에서 제공됨)
#**관성(Inertia)**는 클러스터 내부의 응집도를 측정합니다. 즉, 각 클러스터 내에서 데이터 포인트들이 얼마나 서로 가까운지를 나타냅니다.
# 관성 값이 낮을수록, 데이터 포인트들이 클러스터의 중심에 더 가까운 상태로 모여 있다는 의미입니다. 반대로, 관성 값이 높으면 클러스터들이 더 분산되어 있다는 뜻입니다.
print(f"Inertia: {kmeans.inertia_:.4f}")
```

```python
# Step 4: 토픽 시각화
# 시각화 데이터 준비

lda_vis = pyLDAvis.gensim_models.prepare(lda_model, corpus, dictionary)

# 시각화 디스플레이
pyLDAvis.display(lda_vis)
```

```python
#LDA와 NMF의 결과 해석 방식 차이:

#LDA는 확률 기반으로 단어 간 연관성을 학습하여 각 토픽의 단어 분포를 계산합니다. Coherence 점수는 단어 간의 통계적 동시 출현(co-occurrence)을 기반으로 측정됩니다.
#NMF는 수학적 행렬 분해 방식을 사용하여 토픽을 추출합니다. Coherence 점수는 NMF의 출력이 단어 간 의미적 연관성을 얼마나 잘 포착했는지를 측정합니다.
#두 모델이 본질적으로 서로 다른 방식으로 토픽을 추출하므로, Coherence 점수가 다르게 나옵니다.
```
